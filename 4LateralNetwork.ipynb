```julia

using JuMP
using Gurobi

n = 2;
# number of nodes. substation node not included 
# 7 = 4 + 2 + 1 laterals
N = 7 * n

# number of DGs that can be compromised
M = 10;

# load parameters
PCD = 1/N
QCD = PCD/3

# line impedance parameters
r = 0.01
x = 0.01

# PV parameters
SG = 0.4PCD
PGSP = r/(r^2+x^2)^0.5 * SG
QGSP = r/(r^2+x^2)^0.5 * SG

#root node voltage
nus_0 = 1

# cost of load shedding
Cll = 974
Clovr = 2000 - Cll; 
W = 100Clovr;

currentBestObjValue = -100W;
lovrValue = 0;
vollValue = 0;
targetNodes = zeros(Int64,M);
ngammas = Float64[];

allNodes = [1:N];
for comb in combinations(allNodes, M)
	m = Model();

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	#@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] >= 0)
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)
	@defVar(m, lovr >= 0)
	@defVar(m, voll >= 0)

	# define objective
	obj = 0;
	for i=1:N
		obj += Cll*(1-gammas[i])
	end

	addConstraint(m, voll == obj);
	addConstraint(m, lovr == W*t);

	obj += W*t;

	@setObjective(m, :Min, obj)

	# Constraints
	nusL = 0.97^2;
	nusU = 1.03^2;
	nus0 = 1;

	# forkingNodes
	specialNodes = [n 2n 3n 4n 5n 6n 7n];
	firstLevelForkingNode = n;
	secondLevelForkingNodes = [2n 5n];
	terminalNodes = [3n 4n 6n 7n];

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];

		#voltage constraints
		addConstraint(m, t >= nusL - nus[i]);
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD);
		addConstraint(m, qc[i] == gammas[i]*QCD);

		# PV constraints
		addConstraint(m, pg[i] == deltas[i]*PGSP);
		addConstraint(m, qg[i] == -SG + (SG + QGSP) * deltas[i]);

		# power conservation
		PSUM = pc[i] - pg[i];
		QSUM = qc[i] - qg[i];
		
		if !(i in terminalNodes)
			PSUM += P[i+1];
			QSUM += Q[i+1];		
		end

		if (i == firstLevelForkingNode)
			PSUM += P[4n+1];
			QSUM += Q[4n+1];
		elseif (i in secondLevelForkingNodes)
			PSUM += P[i+n+1];
			QSUM += Q[i+n+1];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]);

		if i == 1
			eqSum += nus0;
		elseif i == 4n+1
			eqSum += nus[n];
		elseif (i == 3n+1 || i == 6n+1)
			eqSum += nus[i-n-1];
		else 
			eqSum += nus[i-1];
		end

		addConstraint(m, nus[i] == eqSum);

	end

	addConstraint(m, deltasSum == N-M);
	for i in comb
		addConstraint(m, deltas[i] == 0);
	end

	@show comb
	status = solve(m);

	objValue = getObjectiveValue(m)+ N*Cll;

	if objValue >= currentBestObjValue
		currentBestObjValue = objValue ;
		lovrValue = getValue(lovr);
		vollValue = getValue(voll);
		ngammas = Float64[];
		j = 1;
		for i = 1:N
			push!(ngammas, getValue(gammas[i]));
			if getValue(deltas[i]) == 0
				targetNodes[j] = i;
				j += 1;
			end
		end
	end
#    break;
end
```
