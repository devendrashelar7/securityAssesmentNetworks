```julia
#Benders' decomposition

using JuMP
using Gurobi

n = 2;
# number of nodes. substation node not included 
# 7 = 4 + 2 + 1 laterals
N = 7 * n

# number of DGs that can be compromised
M = 5;

# line impedance parameters
r = 0.01
x = 0.01

# load parameters
PCD = 1/N
QCD = PCD/3

# cost of load shedding
Cll = 1000;
Clovr = 100*(2000 - Cll);
W = Clovr;

# Constraints
nusL = 0.97^2;
nusU = 1.03^2;
nus0 = 1;

# specify the tree network by defining parents
par = zeros(Int64, N);
for i = 1:N
	par[i] = i - 1;
end
par[3n+1] = 2n;
par[4n+1] = n;
par[6n+1] = 5n;

function getChildren(par)
	child = Array{Int64, 1}[];
	N = length(par);
	for i=1:N
		childi = Int64[];
		for j=1:N
			if par[j] == i
				push!(childi, j);
			end
		end
		push!(child, childi);
	end

	child
end

child = getChildren(par);

currentBestObjValue = -100W;
lovrValue = 0;
vollValue = 0;
targetNodes = zeros(Int64,M);
ngammas = Float64[];

allNodes = [1:N];
for comb in combinations(allNodes, M)
	m = Model();

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, gammas[1:N] >= 0)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	@defVar(m, 0 <= lij[1:N] <= 4)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] >= 0)
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, 0 <= t <= 1)

	# define objective
	obj = 0;
	for i=1:N
		obj += Cll*(1-gammas[i])
	end
	obj += W*t

	@setObjective(m, :Min, obj)

	# Constraints
	nusL = 0.97^2;
	nusU = 1.03^2;
	nus0 = 1;

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		addConstraint(m, gammas[i] <= 1 );

		#voltage constraints
		addConstraint(m, t >= nusL - nus[i]);
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD);
		addConstraint(m, qc[i] == gammas[i]*QCD);

		# PV constraints
		addConstraint(m, pg[i] == deltas[i]*PGSP[i]);
		addConstraint(m, qg[i] == -SG[i] + (SG[i] + QGSP[i]) * deltas[i]);

		# power conservation
		PSUM = pc[i] - pg[i] + r*lij[i];
		QSUM = qc[i] - qg[i] + x*lij[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);


		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) + (r^2 + x^2)*lij[i] ;

		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);

	end

	addConstraint(m, deltasSum == N-M);
	for i in comb
		addConstraint(m, deltas[i] == 0);
	end

	@show comb
	status = solve(m);

	objValue = getObjectiveValue(m);

	if objValue >= currentBestObjValue
		currentBestObjValue = objValue ;
		lovrValue = Clovr * getValue(t);
		vollValue = objValue - lovrValue;
		ngammas = Float64[];
		j = 1;
		for i = 1:N
			push!(ngammas, getValue(gammas[i]));
			if getValue(deltas[i]) == 0
				targetNodes[j] = i;
				j += 1;
			end
		end
	end
end

```
