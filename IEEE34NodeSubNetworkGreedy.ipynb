```julia
#Benders' decomposition

using JuMP
using Gurobi

N = 14;

# number of DGs that can be compromised
M = 5;

# line impedance parameters
r = 0.01
x = 0.01

# load parameters
PCD = 1/N
QCD = PCD/3

# PV parameters
SG = 0.4PCD
PGSP = r/(r^2+x^2)^0.5 * SG
QGSP = r/(r^2+x^2)^0.5 * SG

# cost of load shedding
Cll = 900;
Clovr = 100*(2000 - Cll);
Cll *= PCD/N; 
W = Clovr;

# Constraints
nusL = 0.97^2;
nusU = 1.03^2;
nus0 = 1;

# specify the tree network by defining parents
par = zeros(Int64, N);
for i = 1:N
	par[i] = i - 1;
end
par[7] = 4;
par[9] = 2;
par[13] = 10;

function getChildren(par)
	child = Array{Int64, 1}[];
	N = length(par);
	for i=1:N
		childi = Int64[];
		for j=1:N
			if par[j] == i
				push!(childi, j);
			end
		end
		push!(child, childi);
	end

	child
end

child = getChildren(par);

function countEdgesOnPath(par) 
	nEdgesOnPath = Int64[];
	for i = 1:length(par)
		node = i;
		j = 1;
		while par[node] !=0
			node = par[node];
			j+=1;
		end
		push!(nEdgesOnPath, j);
	end
	nEdgesOnPath
end

nEdgesOnPath = countEdgesOnPath(par);

function getLinearModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	#@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] >= 0)
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD);
		addConstraint(m, qc[i] == gammas[i]*QCD);

		# PV constraints
		addConstraint(m, pg[i] == deltas[i]*PGSP);
		addConstraint(m, qg[i] == -SG + (SG + QGSP) * deltas[i]);

		# power conservation
		PSUM = pc[i] - pg[i];
		QSUM = qc[i] - qg[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) ;
 
		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

	end

	# addConstraint(m, deltasSum == N-M);

	# define objective
	obj = 0;
	for i=1:N
		#obj += C*(1-gammas[i])
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end

function getNonLinearModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] >= 0)
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD);
		addConstraint(m, qc[i] == gammas[i]*QCD);

		# PV constraints
		addConstraint(m, pg[i] == deltas[i]*PGSP);
		addConstraint(m, qg[i] == -SG + (SG + QGSP) * deltas[i]);

		# power conservation
		PSUM = pc[i] - pg[i] + r * lij[i];
		QSUM = qc[i] - qg[i] + x * lij[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) + (r^2 + x^2) * lij[i];
 
		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

		#current constraint
		addConstraint(m, lij[i] * nus[i] >= P[i]^2 + Q[i]^2);

	end

	# addConstraint(m, deltasSum == N-M);

	# define objective
	obj = 0;
	for i=1:N
		obj += Cll*(1-gammas[i]) + r*lij[i]
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end


function countCommonEdges(par)
	paths = Array{Int64,2}[];
	for i = 1:N
		path = Int64[];
		node = i;
		while node != 0;
			push!(path,node);
			node = par[node];
		end
		push!(paths, path');
	end

	commonEdgesList = Array{Int64,2}[];
	for i = 1:N
		commonEdges = zeros(Int64, N);
		p1 = paths[i];
		

		for j = 1:N
			k = 0;	
			p2 = paths[j];
            l1 = length(p1);
			l2 = length(p2);
			while l1 > 0 && l2 > 0 && p1[l1] == p2[l2]
				k += 1;
				l1 -= 1;
				l2 -= 1;
			end
            commonEdges[j] = k;
		end
        push!(commonEdgesList, commonEdges');
	end

    commonEdgesList
end
commonEdgesList = countCommonEdges(par)

objValue = -100Clovr;
lovrValues = Float64[];
vollValues = Float64[];
targetNodesList = Array{Int64,1}[];

for M = 1:N
	# initialize ngammas
	targetNodes = zeros(Int64, M);
	ngammas = ones(Float64, N);
	currentBestObjValue = -100Clovr;
	finalObjValue = -100Clovr;
	ndeltas = zeros(Int64,N);
	maxIterations = 3;
	lovrValue = 0;
	vollValue = 0;


	for iter = 1:maxIterations
		
		################################################################
		# Master problem 
		################################################################
		optimalAttackValues = zeros(Float64, N);
		optimalDeltasForPivotNode = Array{Int64,1}[];
		changeInNu = (r * PGSP + x * (QGSP + SG));

		mm = getLinearModel();
		m = mm[1]; 
		deltas = mm[2];
		gammas = mm[3];
		t = mm[4];
		obj = mm[5];
		nus = mm[6];
		P = mm[7];
		Q = mm[8];
		pc = mm[9];
		qc = mm[10];
		pg = mm[11];
		qg = mm[12];
		
		# if master problem, then minimize the cost for defender
		@setObjective(m, :Min, obj)		
					
		for i = 1:N
			addConstraint(m, gammas[i] == ngammas[i]);				
			addConstraint(m, deltas[i] == 1);
		end

		solve(m);

		for pivotNode = 1:N
			objValues = zeros(Float64, N);
			for targetNode = 1:N
				objValues[targetNode] = commonEdgesList[pivotNode][targetNode] * changeInNu;
			end

			# find optimal target nodes for the pivot node
			optimalDeltas = Int64[];
			optimalAttackValue = getValue(nus[pivotNode]);
			for i = 1:M				
				objValue = -W;
				maxJ = -1;
				for j = 1:N
					if !(j in optimalDeltas) && objValues[j] >= objValue
						objValue = objValues[j];
						maxJ = j;
					end	
				end

				optimalAttackValue -= objValues[maxJ];
				push!(optimalDeltas, maxJ);
			end	

			optimalAttackValues[pivotNode] = optimalAttackValue;
			push!(optimalDeltasForPivotNode, optimalDeltas);
		end	

		# find the best pivot node to target
		minValue = W;
		minI = -1;
		for i = 1:N
			if optimalAttackValues[i] <= minValue
				minI = i;
			end
		end

		optimalTargetNodes = optimalDeltasForPivotNode[minI];
		# initialize optimal attack vector for the best pivot node to target
		for i = 1:N
			if i in optimalTargetNodes
				ndeltas[i] = 0;
			else
				ndeltas[i] = 1;
			end
		end

		################################################################
		# Slave problem 
		################################################################
		mm = getNonLinearModel();
		m = mm[1]; 
		deltas = mm[2];
		gammas = mm[3];
		t = mm[4];
		obj = mm[5];
		nus = mm[6];
		P = mm[7];
		Q = mm[8];
		pc = mm[9];
		qc = mm[10];
		pg = mm[11];
		qg = mm[12];
		
		# if slave problem, then maximize the rewards for attacker
		@setObjective(m, :Min, obj)		
				
		# pivot node voltage constraints
		addConstraint(m, t >= 0);
		for i = 1:N
			addConstraint(m, t >= nusL - nus[i]);
		end	
		
		for i = 1:N	
			addConstraint(m, deltas[i] == ndeltas[i]);
		end

		solve(m)

		ngammas = zeros(Float64, N);
		for i=1:N
		    ngammas[i] = getValue(gammas[i]);
		end

		nnus = zeros(Float64, N);
		for i=1:N
		    nnus[i] = getValue(nus[i]);
		end

		finalObjValue = getObjectiveValue(m);
		if finalObjValue >= currentBestObjValue
			# update values
			currentBestObjValue = finalObjValue;
			lovrValue = Clovr * getValue(t);
	        vollValue = finalObjValue - lovrValue;
	        
			
			j = 1;
			for i=1:N
			    if ndeltas[i] == 0
			    	targetNodes[j] = i;
			    	j += 1;
			    end
			end
		end

	end
	push!(lovrValues, lovrValue);
    push!(vollValues, vollValue);
    push!(targetNodesList, targetNodes);

end
```
