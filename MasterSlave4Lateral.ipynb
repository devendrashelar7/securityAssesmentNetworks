```julia
using JuMP
using Gurobi

n = 2;
# number of nodes. substation node not included 
# 7 = 4 + 2 + 1 laterals
N = 7 * n

# number of DGs that can be compromised
M = 5;

# line impedance parameters
r = 0.01
x = 0.01

# load parameters
PCD = 1/10N
QCD = PCD/3

# PV parameters
SG = 0.4PCD
PGSP = r/(r^2+x^2)^0.5 * SG
QGSP = r/(r^2+x^2)^0.5 * SG

# forkingNodes
specialNodes = [n 2n 3n 4n 5n 6n 7n];
firstLevelForkingNode = n;
secondLevelForkingNodes = [2n 5n];
terminalNodes = [3n 4n 6n 7n];

# cost of load shedding
Cll = 100
Clovr = 100*(2000-Cll);
W = Clovr;

# Constraints
nusL = 0.97^2;
nusU = 1.03^2;
nus0 = 1;

function getLinearModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	#@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] >= 0)
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD);
		addConstraint(m, qc[i] == gammas[i]*QCD);

		# PV constraints
		addConstraint(m, pg[i] == deltas[i]*PGSP);
		addConstraint(m, qg[i] == -SG + (SG + QGSP) * deltas[i]);

		# power conservation
		PSUM = pc[i] - pg[i];
		QSUM = qc[i] - qg[i];
		
		if !(i in terminalNodes)
			PSUM += P[i+1];
			QSUM += Q[i+1];		
		end

		if (i == firstLevelForkingNode)
			PSUM += P[4n+1];
			QSUM += Q[4n+1];
		elseif (i in secondLevelForkingNodes)
			PSUM += P[i+n+1];
			QSUM += Q[i+n+1];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]);

		if i == 1
			eqSum += nus0;
		elseif i == 4n+1
			eqSum += nus[n];
		elseif (i == 3n+1 || i == 6n+1)
			eqSum += nus[i-n-1];
		else 
			eqSum += nus[i-1];
		end

		addConstraint(m, nus[i] == eqSum);

	end

	# addConstraint(m, deltasSum == N-M);

	# define objective
	obj = 0;
	for i=1:N
		#obj += C*(1-gammas[i])
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end

function getNonLinearModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] >= 0)
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD);
		addConstraint(m, qc[i] == gammas[i]*QCD);

		# PV constraints
		addConstraint(m, pg[i] == deltas[i]*PGSP);
		addConstraint(m, qg[i] == -SG + (SG + QGSP) * deltas[i]);

		# power conservation
		PSUM = pc[i] - pg[i] + r*lij[i];
		QSUM = qc[i] - qg[i] + x*lij[i];
		
		if !(i in terminalNodes)
			PSUM += P[i+1];
			QSUM += Q[i+1];		
		end

		if (i == firstLevelForkingNode)
			PSUM += P[4n+1];
			QSUM += Q[4n+1];
		elseif (i in secondLevelForkingNodes)
			PSUM += P[i+n+1];
			QSUM += Q[i+n+1];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) + (r^2+x^2)*lij[i];

		if i == 1
			eqSum += nus0;
		elseif i == 4n+1
			eqSum += nus[n];
		elseif (i == 3n+1 || i == 6n+1)
			eqSum += nus[i-n-1];
		else 
			eqSum += nus[i-1];
		end

		addConstraint(m, nus[i] == eqSum);

		addConstraint(m, lij[i]*nus[i] >= P[i]^2 + Q[i]^2);

	end

	# addConstraint(m, deltasSum == N-M);

	# define objective
	obj = 0;
	for i=1:N
		obj += Cll*(1-gammas[i]) + r*lij[i]
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end

# initialize ngammas
targetNodes = zeros(Int64, M);
ngammas = ones(Float64, N);
currentBestObjValue = -100Clovr;
finalObjValue = -100Clovr;
ndeltas = zeros(Int64,N);
maxIterations = 10;
lovrValue = 0;
vollValue = 0;


for iter = 1:maxIterations
	
	################################################################
	# Master problem 
	################################################################
	optimalAttackValues = zeros(Float64, N);
	optimalDeltasForPivotNode = Array{Int64,1}[];
	for pivotNode = 1:N
		objValues = zeros(Float64, N);
		for targetNode = 1:N
			mm = getLinearModel();
			m = mm[1]; 
			deltas = mm[2];
			gammas = mm[3];
			t = mm[4];
			obj = mm[5];
			nus = mm[6];
			P = mm[7];
			Q = mm[8];
			pc = mm[9];
			qc = mm[10];
			pg = mm[11];
			qg = mm[12];
			
			# if master problem, then minimize the cost for defender
			@setObjective(m, :Max, obj)		
						
			#voltage constraints
			addConstraint(m, t == nusL - nus[pivotNode]);					
			for i = 1:N
				addConstraint(m, gammas[i] == ngammas[i]);				
				if i == targetNode			
					addConstraint(m, deltas[i] == 0);
				else
					addConstraint(m, deltas[i] == 1);
				end
			end

			solve(m)

			objValues[targetNode] = getObjectiveValue(m);

		end

		# find optimal target nodes for the pivot node
		optimalDeltas = Int64[];
		optimalAttackValue = 0;
		for i = 1:M				
			objValue = -W;
			maxJ = -1;
			for j = 1:N
				if !(j in optimalDeltas) && objValues[j] > objValue
					objValue = objValues[j];
					maxJ = j;
				end	
			end

			optimalAttackValue += objValues[maxJ];
			push!(optimalDeltas, maxJ);
		end	

		optimalAttackValues[pivotNode] = optimalAttackValue;
		push!(optimalDeltasForPivotNode, optimalDeltas);
	end	

	# find the best pivot node to target
	maxValue = -W;
	maxI = -1;
	for i = 1:N
		if optimalAttackValues[i] > maxValue
			maxI = i;
		end
	end

	optimalTargetNodes = optimalDeltasForPivotNode[maxI];
	# initialize optimal attack vector for the best pivot node to target
	for i = 1:N
		if i in optimalTargetNodes
			ndeltas[i] = 0;
		else
			ndeltas[i] = 1;
		end
	end

	nnus = zeros(Float64, N);
	for i=1:N
	    nnus[i] = getValue(nus[i]);
	end

	################################################################
	# Slave problem 
	################################################################
	mm = getNonLinearModel();
	m = mm[1]; 
	deltas = mm[2];
	gammas = mm[3];
	t = mm[4];
	obj = mm[5];
	nus = mm[6];
	P = mm[7];
	Q = mm[8];
	pc = mm[9];
	qc = mm[10];
	pg = mm[11];
	qg = mm[12];
	
	# if slave problem, then maximize the rewards for attacker
	@setObjective(m, :Min, obj)		
			
	# pivot node voltage constraints
	addConstraint(m, t >= 0);
	for i = 1:N
		addConstraint(m, t >= nusL - nus[i]);
	end	
	
	for i = 1:N	
		addConstraint(m, deltas[i] == ndeltas[i]);
	end

	solve(m)

	ngammas = zeros(Float64, N);
	for i=1:N
	    ngammas[i] = getValue(gammas[i]);
	end

	nnus = zeros(Float64, N);
	for i=1:N
	    nnus[i] = getValue(nus[i]);
	end

	finalObjValue = getObjectiveValue(m) + N*Cll;
	if finalObjValue >= currentBestObjValue
		# update values
		currentBestObjValue = finalObjValue;
		lovrValue = Clovr * getValue(t);
        vollValue = finalObjValue - lovrValue;
		
		j = 1;
		for i=1:N
		    if ndeltas[i] == 0
		    	targetNodes[j] = i;
		    	j += 1;
		    end
		end
	end
end


```
