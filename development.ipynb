```julia
using JuMP;

m = Model(solver=GurobiSolver());

#number of nodes. substation node not included
N=4

# Binary Variables
@defVar(m, deltas[1:N], Bin)
@defVar(m, gammas[1:N], Bin)

# voltage and current magnitude squares
@defVar(m, nu[1:N] >=0, Float64)
@defVar(m, lij[1:N] >=0, Float64)

# real and reactive and complex loads and DGs
@defVar(m, pc[1:N] >=0, Float64)
@defVar(m, qc[1:N] >=0, Float64)
@defVar(m, pg[1:N] >=0, Float64)
@defVar(m, qg[1:N] >=0, Float64)
@defVar(m, sg[1:N] >=0, Float64)

# real and reactive power flows
@defVar(m, P[1:N] >=0, Float64)
@defVar(m, Q[1:N] >=0, Float64)

# power parameters
PCL = 0.9
QCL = 0.3
PGM = 0.4
SGM = 0.44
nu_0 = 1

# line impedance parameters
r = 0.15
x = 0.15

# cost of load shedding
C = 1000

# define objective
obj = 0;
for i=1:N
	obj += r * lij[i] + C(1-gammas[i])
end
@setObjective(m, :Min, obj)


# Constraints
VL = 0.97;
VU = 1.03;

# Number of DGs that can be compromised
M = 2;

deltasSum = 0;
for i=1:N
	addConstraint(m, VL <= nu[i] );
	addConstraint(m, nu[i] <= VM );
	addConstraint(m, PCL <= pc[i] );
	addConstraint(m, QCL <= qc[i] );
	addConstraint(m, sg[i] <= SGM * deltas[i] );
	addConstraint(m, sg[i]^2 >= pg[i]^2 + qg[i]^2 );

	if 2 <= i && i < N 
	addConstraint(m, nu[i] == nu[i-1] - 2(r * P[i] + x * Q[i]) + (r^2 + x^2) * lij[i] == 0);

	# sums
	deltasSum += deltas[i];
	PSum = r * lij[i] + pc[i] - pg[i];
	QSum = x * lij[i] + qc[i] - qg[i];
	
	for j=i+1:N
		PSum += P[j];
		QSum += Q[j];
	end	

	addConstraint(m, P[i] == PSum);
	addConstraint(m, Q[i] == QSum);
end

addConstraint(m, deltasSum == N - M );
addConstraint(m, nu[1] - nu_0 + 2(r * P[1] + x * Q[1]) - (r^2 + x^2) * lij[1] == 0);

m
```
