```julia
using JuMP
using Gurobi
N = 36

function getChildren(par)
	child = Array{Int64, 1}[];
	N = length(par);
	for i=1:N
		childi = Int64[];
		for j=1:N
			if par[j] == i
				push!(childi, j);
			end
		end
		push!(child, childi);
	end

	child
end

function countEdgesOnPath(par) 
	nEdgesOnPath = Int64[];
	for i = 1:length(par)
		node = i;
		j = 1;
		while par[node] !=0
			node = par[node];
			j+=1;
		end
		push!(nEdgesOnPath, j);
	end
	nEdgesOnPath
end

function countCommonEdges(par)
	paths = Array{Int64,2}[];
	for i = 1:N
		path = Int64[];
		node = i;
		while node != 0;
			push!(path,node);
			node = par[node];
		end
		push!(paths, path');
	end

	commonEdgesList = Array{Int64,2}[];
	for i = 1:N
		commonEdges = zeros(Int64, N);
		p1 = paths[i];
		

		for j = 1:N
			k = 0;	
			p2 = paths[j];
            l1 = length(p1);
			l2 = length(p2);
			while l1 > 0 && l2 > 0 && p1[l1] == p2[l2]
				k += 1;
				l1 -= 1;
				l2 -= 1;
			end
            commonEdges[j] = k;
		end
        push!(commonEdgesList, commonEdges');
	end

    commonEdgesList
end


function getLinearModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	#@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] )
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD[i]);
		addConstraint(m, qc[i] == gammas[i]*QCD[i]);

				# PV constraints
        addConstraint(m, pg[i] <= deltas[i]*SGmax[i]);
        addConstraint(m, qg[i] <= -SGmax[i] + 2 * SGmax[i] * deltas[i]);
        addConstraint(m, qg[i] >= -SGmax[i]);
        addConstraint(m, pg[i]^2 + qg[i]^2 <= SGmax[i]^2);

		# power conservation
		PSUM = pc[i] - pg[i];
		QSUM = qc[i] - qg[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) ;
 
		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

	end

	# addConstraint(m, deltasSum == N-M);

	# define objective
	obj = 0;
	for i=1:N
		#obj += Cll * (1-gammas[i]) * PCD[i]
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end

function getNonLinearModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, gammaMin <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] )
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD[i]);
		addConstraint(m, qc[i] == gammas[i]*QCD[i]);

				# PV constraints
        addConstraint(m, pg[i] <= deltas[i]*SGmax[i]);
        addConstraint(m, qg[i] <= -SGmax[i] + 2 * SGmax[i] * deltas[i]);
        addConstraint(m, qg[i] >= -SGmax[i]);
        addConstraint(m, pg[i]^2 + qg[i]^2 <= SGmax[i]^2);

		# power conservation
		PSUM = pc[i] - pg[i] + r * lij[i];
		QSUM = qc[i] - qg[i] + x * lij[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) + (r^2 + x^2) * lij[i];
 
		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

		#current constraint
		addConstraint(m, lij[i] * nus[i] >= P[i]^2 + Q[i]^2);

	end

	# addConstraint(m, deltasSum == N-M);

	# define objective
	obj = 0;
	for i=1:N
		obj += Cll * (1-gammas[i]) * PCD[i] + r*lij[i]
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end
```

```julia
Cll = 6
Clovr = 100000
objValue = -100Clovr;
lovrValues = Float64[];
vollValues = Float64[];
targetNodesList = Array{Int64,1}[];
nnus = zeros(Float64, N);

#for Cllp = 3:7
for Cllp = 6:6
    
    Clovrp = 10 - Cllp;
    Cll = Cllp/1000 * Sbase;
    Clovr = Clovrp * Vbase / 2;
    W = Clovr;
    #for M = 5:10
    for M = 8:8
        # initialize ngammas
        targetNodes = zeros(Int64, M);
        ngammas = ones(Float64, N);
        ndeltas = ones(Float64, N);
        currentBestObjValue = -100Clovr;
        finalObjValue = -100Clovr;
        ndeltas = zeros(Int64,N);
        maxIterations = 10;
        lovrValue = 0;
        vollValue = 0;
        tempTargetNodesList = Array{Int64,1}[];

        for iter = 1:maxIterations

            ################################################################
            # Master problem 
            ################################################################
            optimalAttackValues = zeros(Float64, N);
            optimalDeltasForPivotNode = Array{Int64,1}[];

            mm = getLinearModel();
            m = mm[1]; 
            deltas = mm[2];
            gammas = mm[3];
            t = mm[4];
            obj = mm[5];
            nus = mm[6];
            P = mm[7];
            Q = mm[8];
            pc = mm[9];
            qc = mm[10];
            pg = mm[11];
            qg = mm[12];

            # if master problem, then minimize the cost for defender
            @setObjective(m, :Min, obj)		

            for i = 1:N
                addConstraint(m, gammas[i] == ngammas[i]);				
                addConstraint(m, deltas[i] == 1);
            end

            solve(m);

            for pivotNode = 1:N

                objValues = zeros(Float64, N);
                for targetNode = 1:N
                    changeInNu = (r * PGSP[targetNode] + x * (QGSP[targetNode] + SGmax[targetNode]));
                    objValues[targetNode] = commonEdgesList[pivotNode][targetNode] * changeInNu;
                end

                # find optimal target nodes for the pivot node
                optimalDeltas = Int64[];
                optimalAttackValue = getValue(nus[pivotNode]);
                for i = 1:M				
                    objValue = -W;
                    maxJ = -1;
                    for j = 1:N
                        if !(j in optimalDeltas) && objValues[j] >= objValue
                            objValue = objValues[j];
                            maxJ = j;
                        end	
                    end

                    optimalAttackValue -= objValues[maxJ];
                    push!(optimalDeltas, maxJ);
                end	

                optimalAttackValues[pivotNode] = optimalAttackValue;
                push!(optimalDeltasForPivotNode, optimalDeltas);
            end	

            # find the best pivot node to target
            minValue = W;
            minI = -1;
            for i = 1:N
                if optimalAttackValues[i] <= minValue
                    minI = i;
                end
            end

            optimalTargetNodes = optimalDeltasForPivotNode[minI];
            # initialize optimal attack vector for the best pivot node to target
            for i = 1:N
                if i in optimalTargetNodes
                    ndeltas[i] = 0;
                else
                    ndeltas[i] = 1;
                end
            end

            ################################################################
            # Slave problem 
            ################################################################
            mm = getNonLinearModel();
            m = mm[1]; 
            deltas = mm[2];
            gammas = mm[3];
            t = mm[4];
            obj = mm[5];
            nus = mm[6];
            P = mm[7];
            Q = mm[8];
            pc = mm[9];
            qc = mm[10];
            pg = mm[11];
            qg = mm[12];

            # if slave problem, then maximize the rewards for attacker
            @setObjective(m, :Min, obj)		

            # pivot node voltage constraints
            addConstraint(m, t >= 0);
            for i = 1:N
                addConstraint(m, t >= nusL - nus[i]);
            end	

            for i = 1:N	
                addConstraint(m, deltas[i] == ndeltas[i]);
                #addConstraint(m, deltas[i] == 1);
                #addConstraint(m, gammas[i] == 1);
            end

            solve(m)

            ngammas = zeros(Float64, N);
            for i=1:N
                ngammas[i] = getValue(gammas[i]);
            end

            nnus = zeros(Float64, N);
            for i=1:N
                nnus[i] = getValue(nus[i]);
            end

            finalObjValue = getObjectiveValue(m);
            if finalObjValue >= currentBestObjValue
                # update values
                currentBestObjValue = finalObjValue;
                lovrValue = Clovr * getValue(t);
                vollValue = 0;
                for i = 1:N
                    vollValue += Cll*(1 - ngammas[i]) * PCD[i];
                end

                j = 1;
                for i=1:N
                    if ndeltas[i] == 0
                        targetNodes[j] = i;
                        j += 1;
                    end
                end

                if targetNodes in tempTargetNodesList
                    @show targetNodes
                    break;
                else
                    push!(tempTargetNodesList, targetNodes);
                end
            end
        end

        push!(lovrValues, lovrValue);
        push!(vollValues, vollValue);
        push!(targetNodesList, targetNodes);

    end
end
```

```julia
function getADLP2Model()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:ny], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	#@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] >= 0)
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] >= 0)

	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	for i=1:N
		addConstraint(m, deltas[i] == deltas[i+N] );
		#addConstraint(m, deltas[i+2N] == deltas[i+N] );
		
		# load constraints
        addConstraint(m, pc[i] == gammas[i]*PCD[i]);
        addConstraint(m, qc[i] == gammas[i]*QCD[i]);

		# PV constraints
		# addConstraint(m, pg[i] == (1 - deltas[i]) * PGSP);
        addConstraint(m, pg[i] <= PGSP[i]);
        addConstraint(m, qg[i] <= QGSP[i] + SGmax[i]);
        #addConstraint(m, qg[i] >= 0);

		# power conservation
		PSUM = pc[i] - pg[i];
        QSUM = qc[i] - qg[i] + SGmax[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]);

		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

	end

	# deltas constraints
    for i = 1:ny
		addConstraint(m, deltas[i] == ndeltas[i]);
	end

	# define cost vector c and defense y vector, and attack vector d
	c = zeros(Float64,ny);
	for i = 7N+1:8N
        c[i] = Cll * PCD[i-7N];
	end
	c[ny] = -Clovr;

	y = typeof(t)[];
	for i = 1:ny
		if i <= N
			push!(y, pg[i]);
		elseif i <= 2N 
			push!(y, qg[i-N]);
		elseif i <= 3N 
			push!(y, pc[i-2N]);
		elseif i <= 4N 
			push!(y, qc[i-3N]);
		elseif i <= 5N 
			push!(y, P[i-4N]);
		elseif i <= 6N 
			push!(y, Q[i-5N]);
		elseif i <= 7N 
			push!(y, nus[i-6N]);
		elseif i <= 8N 
			push!(y, gammas[i-7N]);
		else
			push!(y, t);
		end
	end

	# define objective
	cc = zeros(Float64, ny); 
	for i = 1:ny
		cc[i] = (c[i] - ndeltas[i] * rbar[i]) ;
	end

	obj = sum (cc' * y);
    #@show obj

	m, deltas, gammas, t, obj, nus, P, Q, pc, qc, pg, qg, c, y, cc
end
```

```julia
function getLinearFreeModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	#@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] )
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD[i]);
		addConstraint(m, qc[i] == gammas[i]*QCD[i]);

				# PV constraints
        addConstraint(m, pg[i] <= deltas[i]*PGSP[i]);
        addConstraint(m, qg[i] <= -SGmax[i] + (SGmax[i] + QGSP[i]) * deltas[i]);
        addConstraint(m, qg[i] >= -SGmax[i]);

		# power conservation
		PSUM = pc[i] - pg[i];
		QSUM = qc[i] - qg[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) ;
 
		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

	end

	# addConstraint(m, deltasSum == N-M);

	# define objective
	obj = 0;
	for i=1:N
		#obj += Cll * (1-gammas[i]) * PCD[i]
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end
function getNonLinearFreeModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] )
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD[i]);
		addConstraint(m, qc[i] == gammas[i]*QCD[i]);

				# PV constraints
        addConstraint(m, pg[i] <= deltas[i]*PGSP[i]);
        addConstraint(m, qg[i] <= -SGmax[i] + (SGmax[i] + QGSP[i]) * deltas[i]);
        addConstraint(m, qg[i] >= -SGmax[i]);

		# power conservation
		PSUM = pc[i] - pg[i] + r * lij[i];
		QSUM = qc[i] - qg[i] + x * lij[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) + (r^2 + x^2) * lij[i];
 
		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

		#current constraint
		addConstraint(m, lij[i] * nus[i] >= P[i]^2 + Q[i]^2);

	end

	# addConstraint(m, deltasSum == N-M);

	# define objective
	obj = 0;
	for i=1:N
		obj += Cll * (1-gammas[i]) * PCD[i] + r*lij[i]
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end
```
