```julia
#Benders' decomposition

using JuMP
using Gurobi

n = 1;
# number of nodes. substation node not included 
# 7 = 4 + 2 + 1 laterals
N = 7 * n

# number of DGs that can be compromised
M = 3;

# line impedance parameters
r = 0.01
x = 0.01

# load parameters
PCD = 2/N
QCD = PCD/3

# PV parameters
SG = 0.4PCD
PGSP = r/(r^2+x^2)^0.5 * SG
QGSP = r/(r^2+x^2)^0.5 * SG

# cost of load shedding
Cll = 1321
Clovr = 100*(2000-Cll)
W = Clovr

# Constraints
nusL = 0.97^2;
nusU = 1.03^2;
nus0 = 1;

# specify the tree network by defining parents
par = zeros(Int64, N);
for i = 1:N
	par[i] = i - 1;
end
par[3n+1] = 2n;
par[4n+1] = n;
par[6n+1] = 5n;

function getChildren(par)
	child = Array{Int64, 1}[];
	N = length(par);
	for i=1:N
		childi = Int64[];
		for j=1:N
			if par[j] == i
				push!(childi, j);
			end
		end
		push!(child, childi);
	end

	child
end

child = getChildren(par);

function countEdgesOnPath(par) 
	nEdgesOnPath = Int64[];
	for i = 1:length(par)
		node = i;
		j = 1;
		while par[node] !=0
			node = par[node];
			j+=1;
		end
		push!(nEdgesOnPath, j);
	end
	nEdgesOnPath
end

nEdgesOnPath = countEdgesOnPath(par);

ny = 8N+1;

function getRBar()
	diameter = maximum(nEdgesOnPath);
	R = zeros(Float64, ny);
	for i = 1:ny
		if i <= N
			# pg 
			R[i] = 2 * nEdgesOnPath[i] * r * Clovr; 
		elseif i <= 2N
			# qg
			R[i] = 2 * nEdgesOnPath[i-N] * x * Clovr; # qg
		elseif i <= 3N
			# pc
			R[i] = -2 * nEdgesOnPath[i-2N] * r * Clovr;
		elseif i <= 4N
			# qc 
			R[i] = -2 * nEdgesOnPath[i-3N] * x * Clovr;
		elseif i <= 5N
			# P 
			# R[i] = -2 * diameter * r * Clovr;
			R[i] = 0;
		elseif i <= 6N
			# Q
			# R[i] = -2 * diameter * x * Clovr;
			R[i] = 0;
		elseif i <= 7N
			# nu
			# R[i] = Clovr;
			R[i] = 0;
		elseif i <= 8N
			# gamma
			R[i] = Cll - 2 * (r * PCD + x * QCD);
		else 
			# t
			R[i] = -Clovr;
		end
	end
	R
end

rbar = getRBar();

function getADLP2Model()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:ny], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	#@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] >= 0)
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] >= 0)

	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	for i=1:N
		addConstraint(m, deltas[i] == deltas[i+N] );
		addConstraint(m, deltas[i+2N] == deltas[i+N] );
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD);
		addConstraint(m, qc[i] == gammas[i]*QCD);

		# PV constraints
		# addConstraint(m, pg[i] == (1 - deltas[i]) * PGSP);
		addConstraint(m, pg[i] <= PGSP);
		addConstraint(m, qg[i] <= QGSP + SG);

		# power conservation
		PSUM = pc[i] - pg[i];
		QSUM = qc[i] - qg[i] + SG;
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]);

		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

	end

	# deltas constraints
	for i = 1:N
		addConstraint(m, deltas[i] == ndeltas[i]);
	end

	# define cost vector c and defense y vector, and attack vector d
	c = zeros(Float64,ny);
	for i = 7N:8N
		c[i] = Cll;
	end
	c[ny] = -Clovr;

	y = typeof(t)[];
	for i = 1:ny
		if i <= N
			push!(y, pg[i]);
		elseif i <= 2N 
			push!(y, qg[i-N]);
		elseif i <= 3N 
			push!(y, pc[i-2N]);
		elseif i <= 4N 
			push!(y, qc[i-3N]);
		elseif i <= 5N 
			push!(y, P[i-4N]);
		elseif i <= 6N 
			push!(y, Q[i-5N]);
		elseif i <= 7N 
			push!(y, nus[i-6N]);
		elseif i <= 8N 
			push!(y, gammas[i-7N]);
		else
			push!(y, t);
		end
	end

	# define objective
	cc = zeros(Float64, ny); 
	for i = 1:ny
		cc[i] = (c[i] - ndeltas[i] * rbar[i]) ;
	end

	obj = sum (cc' * y);

	m, deltas, gammas, t, obj, nus, P, Q, pc, qc, pg, qg, c, y, cc
end

targetNodes = zeros(Int64, M);
targetNodesStar = zeros(Int64, M);
ndeltas = zeros(Int64, ny);
ndeltaStar = zeros(Int64, ny);
zL = -(Clovr + N * Cll);
zU = -zL;
zhat = 0;
Y = Array{Float64, 1}[];

maxIterations = 36;

epsilon = 1;

# master problem ADMP2
mp = Model();
@defVar(mp, z);
@defVar(mp, mdeltas[1:ny], Bin);
d = typeof(z)[];
for i = 1:N
	push!(d, mdeltas[i]);
end

addConstraint(mp, sum(d) == M);
# delta constraints
for i = 1:N
	addConstraint(mp, mdeltas[i] == mdeltas[i+N]);
end
for i = 2N + 1:ny
	addConstraint(mp, mdeltas[i] == 0);
end

@setObjective(mp, :Min, z);

```

```julia
m = 0
lovrValues = Float64[];
vollValues = Float64[];
targetNodesList = Array{Int64,1}[];
lovrValue = 0;
vollValue = 0;


for iter = 1:maxIterations
    @show iter
	# slave problem
	m, deltas, gammas, t, obj, nus, P, Q, pc, qc, pg, qg, c, y, cc = getADLP2Model();
	
	# if master problem, then minimize the cost for defender
	@setObjective(m, :Max, obj);		

	solve(m)

	y = Float64[];
	for i = 1:ny
		if i <= N
			push!(y, getValue(pg[i]));
		elseif i <= 2N 
			push!(y, getValue(qg[i-N]));
		elseif i <= 3N 
			push!(y, getValue(pc[i-2N]));
		elseif i <= 4N 
			push!(y, getValue(qc[i-3N]));
		elseif i <= 5N 
			push!(y, getValue(P[i-4N]));
		elseif i <= 6N 
			push!(y, getValue(Q[i-5N]));
		elseif i <= 7N 
			push!(y, getValue(nus[i-6N]));
		elseif i <= 8N 
			push!(y, getValue(gammas[i-7N]));
		else
			push!(y, getValue(t));
		end
	end

	zhat = getObjectiveValue(m);

	push!(Y, y);
	if zhat < zU 
		zU = zhat;
		j = 1;
		for i = 1:ny
			ndeltaStar[i] = ndeltas[i];
			if i <= N && ndeltas[i] == 1
				targetNodesStar[j] = i;
				j += 1;
			end	
		end
        
        lovrValue = Clovr*getValue(t);
        vollValue = 0;
        for i = 1:N
            vollValue += Cll*(1 - getValue(gammas[i]))*PCD;
        end
        
        @show targetNodesStar
	end

	if zU - zL <= epsilon
		break;
	end

	# master problem
	lhs = z;
	for i = 1:ny
		lhs += y[i] * rbar[i] * mdeltas[i];
	end
	addConstraint(mp, lhs >= sum(c' * y));
	solve(mp);

	z2 = getObjectiveValue(mp);
	zL = z2;
	j = 1;
	for i=1:ny
		ndeltas[i] = getValue(mdeltas[i]);
		if i <= N && ndeltas[i] == 1
			targetNodes[j] = i;
			j += 1;
		end
	end

	if zU - zL <= epsilon
		break;
	end

end

push!(lovrValues, lovrValue);
push!(vollValues, vollValue);
```
