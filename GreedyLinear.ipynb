```julia
#Benders' decomposition

using JuMP
using Gurobi

n = 2;
# number of nodes. substation node not included 
# 7 = 4 + 2 + 1 laterals
N = 7 * n

# number of DGs that can be compromised
M = 5;

# line impedance parameters
r = 0.01
x = 0.01

# load parameters
PCD = 1/N
QCD = PCD/3

# PV parameters
SG = 0.4PCD
PGSP = r/(r^2+x^2)^0.5 * SG
QGSP = r/(r^2+x^2)^0.5 * SG

# cost of load shedding
Cll = 900;
Clovr = 100*(2000 - Cll);
W = Clovr;

# Constraints
nusL = 0.97^2;
nusU = 1.03^2;
nus0 = 1;

# specify the tree network by defining parents
par = zeros(Int64, N);
for i = 1:N
	par[i] = i - 1;
end
par[3n+1] = 2n;
par[4n+1] = n;
par[6n+1] = 5n;


child = getChildren(par);



nEdgesOnPath = countEdgesOnPath(par);



commonEdgesList = countCommonEdges(par)

objValue = -100Clovr;
lovrValues = Float64[];
vollValues = Float64[];
targetNodesList = Array{Int64,1}[];

for M = 1:14
	# initialize ngammas
	targetNodes = zeros(Int64, M);
	ngammas = ones(Float64, N);
	currentBestObjValue = -100Clovr;
	finalObjValue = -100Clovr;
	ndeltas = zeros(Int64,N);
	maxIterations = 3;
	lovrValue = 0;
	vollValue = 0;


	for iter = 1:maxIterations
		
		################################################################
		# Master problem 
		################################################################
		optimalAttackValues = zeros(Float64, N);
		optimalDeltasForPivotNode = Array{Int64,1}[];
		changeInNu = (r * PGSP + x * (QGSP + SG));

		mm = getLinearModel();
		m = mm[1]; 
		deltas = mm[2];
		gammas = mm[3];
		t = mm[4];
		obj = mm[5];
		nus = mm[6];
		P = mm[7];
		Q = mm[8];
		pc = mm[9];
		qc = mm[10];
		pg = mm[11];
		qg = mm[12];
		
		# if master problem, then minimize the cost for defender
		@setObjective(m, :Min, obj)		
					
		for i = 1:N
			addConstraint(m, gammas[i] == ngammas[i]);				
			addConstraint(m, deltas[i] == 1);
		end

		solve(m);

		for pivotNode = 1:N
			objValues = zeros(Float64, N);
			for targetNode = 1:N
				objValues[targetNode] = commonEdgesList[pivotNode][targetNode] * changeInNu;
			end

			# find optimal target nodes for the pivot node
			optimalDeltas = Int64[];
			optimalAttackValue = getValue(nus[pivotNode]);
			for i = 1:M				
				objValue = -W;
				maxJ = -1;
				for j = 1:N
					if !(j in optimalDeltas) && objValues[j] >= objValue
						objValue = objValues[j];
						maxJ = j;
					end	
				end

				optimalAttackValue -= objValues[maxJ];
				push!(optimalDeltas, maxJ);
			end	

			optimalAttackValues[pivotNode] = optimalAttackValue;
			push!(optimalDeltasForPivotNode, optimalDeltas);
		end	

		# find the best pivot node to target
		minValue = W;
		minI = -1;
		for i = 1:N
			if optimalAttackValues[i] <= minValue
				minI = i;
			end
		end

		optimalTargetNodes = optimalDeltasForPivotNode[minI];
		# initialize optimal attack vector for the best pivot node to target
		for i = 1:N
			if i in optimalTargetNodes
				ndeltas[i] = 0;
			else
				ndeltas[i] = 1;
			end
		end

		################################################################
		# Slave problem 
		################################################################
		mm = getNonLinearModel();
		m = mm[1]; 
		deltas = mm[2];
		gammas = mm[3];
		t = mm[4];
		obj = mm[5];
		nus = mm[6];
		P = mm[7];
		Q = mm[8];
		pc = mm[9];
		qc = mm[10];
		pg = mm[11];
		qg = mm[12];
		
		# if slave problem, then maximize the rewards for attacker
		@setObjective(m, :Min, obj)		
				
		# pivot node voltage constraints
		addConstraint(m, t >= 0);
		for i = 1:N
			addConstraint(m, t >= nusL - nus[i]);
		end	
		
		for i = 1:N	
			addConstraint(m, deltas[i] == ndeltas[i]);
		end

		solve(m)

		ngammas = zeros(Float64, N);
		for i=1:N
		    ngammas[i] = getValue(gammas[i]);
		end

		nnus = zeros(Float64, N);
		for i=1:N
		    nnus[i] = getValue(nus[i]);
		end

		finalObjValue = getObjectiveValue(m);
		if finalObjValue >= currentBestObjValue
			# update values
			currentBestObjValue = finalObjValue;
			lovrValue = Clovr * getValue(t);
	        vollValue = finalObjValue - lovrValue;
	        
			
			j = 1;
			for i=1:N
			    if ndeltas[i] == 0
			    	targetNodes[j] = i;
			    	j += 1;
			    end
			end
		end

	end
	push!(lovrValues, lovrValue);
    push!(vollValues, vollValue);
    push!(targetNodesList, targetNodes);

end
```
