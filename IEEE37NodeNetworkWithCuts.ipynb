```julia
using JuMP
using Gurobi

N = 36;

# number of DGs that can be compromised
M = 5;

#Base Values
PDmax = 5000;
Pbase = N * PDmax;
Sbase = Pbase * (1+0.3^2)^0.5;
Vbase = 2400;

#average length of lines
lineLength = 1000; # 1000 ft

# line impedance parameters
r = 0.33*lineLength/5280;
x = 0.38*lineLength/5280;

# load parameters
PCDmax = PDmax/Sbase;
PCDS = zeros(Float64, N);
QCDS = zeros(Float64, N);
for i = 1:N
	PCDS[i] = PCDmax * (1 - 0.1 * rand());
	QCDS[i] = PCDS[i] * (0.2 + 0.1 * rand());
end

# PV parameters
SGmaxS = zeros(Float64, N);
PGmaxS = zeros(Float64, N);
PGSPS = zeros(Float64, N);
QGSPS = zeros(Float64, N);

for i = 1:N
        random = rand();
    if random < 0.333
        PGmaxS[i] = 0.5PDmax/Sbase;
    elseif random < 0.666
        PGmaxS[i] = 0.6PDmax/Sbase;
    else
        PGmaxS[i] = 0.7PDmax/Sbase;
    end

    SGmaxS[i] = 1.1PGmaxS[i];
    PGSPS[i] = r/(r^2+x^2)^0.5 * SGmaxS[i];
    QGSPS[i] = r/(r^2+x^2)^0.5 * SGmaxS[i];
end

SGmax = zeros(Float64, N);
PGmax = zeros(Float64, N);
PGSP = zeros(Float64, N);
QGSP = zeros(Float64, N);
PCD = zeros(Float64, N);
QCD = zeros(Float64, N);

for i = 1:N
    SGmax[i] = SGmaxS[i];
    PGmax[i] = PGmaxS[i];
    PGSP[i] = PGSPS[i];
    QGSP[i] = QGSPS[i];
    PCD[i] = PCDS[i];
    QCD[i] = QCDS[i];
end

# cost of load shedding
Cllp = 7;
Clovrp = 8 - Cllp;
Cll = 7/1000 * Sbase;
Clovr = Clovrp/1000 * Vbase^2;
W = Clovr;

# Constraints
nusL = 0.95^2;
nusU = 1.05^2;
nus0 = 1;

# specify the tree network by defining parents
par = zeros(Int64, N);
for i = 1:N
	par[i] = i - 1;
end
par[5] = 3;
par[6] = 2;
par[10] = 7;
par[13] = 11;
par[14] = 10;
par[16] = 2;
par[20] = 18;
par[21] = 16;
par[24] = 22;
par[25] = 22;
par[27] = 25;
par[31] = 29;
par[32] = 28;
par[36] = 34;

function getChildren(par)
	child = Array{Int64, 1}[];
	N = length(par);
	for i=1:N
		childi = Int64[];
		for j=1:N
			if par[j] == i
				push!(childi, j);
			end
		end
		push!(child, childi);
	end

	child
end

child = getChildren(par);

function countEdgesOnPath(par) 
	nEdgesOnPath = Int64[];
	for i = 1:length(par)
		node = i;
		j = 1;
		while par[node] !=0
			node = par[node];
			j+=1;
		end
		push!(nEdgesOnPath, j);
	end
	nEdgesOnPath
end

nEdgesOnPath = countEdgesOnPath(par);

function determineAncestry(par)
	ancestors = zeros(Int64, N, N);
	for i = 1:N
		j = i;
		while j !=0
			ancestors[i,j] = 1;
			j = par[j];
		end
	end
	ancestors
end

ancestors = determineAncestry(par);

function getNonLinearModel()
	
	m = Model(solver=GurobiSolver());

	# Binary Variables
	@defVar(m, deltas[1:N], Bin)

	# continous variables
	@defVar(m, 0 <= gammas[1:N] <= 1)

	# voltage and current magnitude squares
	@defVar(m, nus[1:N] >= 0)
	@defVar(m, lij[1:N] >= 0)

	# real and reactive and complex loads and DGs
	@defVar(m, pc[1:N] >= 0)
	@defVar(m, qc[1:N] )
	@defVar(m, pg[1:N] >= 0)
	@defVar(m, qg[1:N] )


	# real and reactive power flows
	@defVar(m, P[1:N] )
	@defVar(m, Q[1:N] )

	#auxiliary variable
	@defVar(m, t >= 0)

	deltasSum = 0;

	for i=1:N
		deltasSum += deltas[i];
		
		# load constraints
		addConstraint(m, pc[i] == gammas[i]*PCD[i]);
		addConstraint(m, qc[i] == gammas[i]*QCD[i]);

				# PV constraints
        addConstraint(m, pg[i] == deltas[i]*PGSP[i]);
        addConstraint(m, qg[i] == -SGmax[i] + (SGmax[i] + QGSP[i]) * deltas[i]);

		# power conservation
		PSUM = pc[i] - pg[i] + r * lij[i];
		QSUM = qc[i] - qg[i] + x * lij[i];
		
		for j in child[i]
			PSUM += P[j];
			QSUM += Q[j];
		end

		addConstraint(m, P[i] == PSUM);
		addConstraint(m, Q[i] == QSUM);

		#ohms law
		eqSum = -2*(r*P[i] + x*Q[i]) + (r^2 + x^2) * lij[i];
 
		if i == 1
			eqSum += nus0;
		else
			eqSum += nus[par[i]];
		end

		addConstraint(m, nus[i] == eqSum);
		addConstraint(m, t >= nusL - nus[i]);

		#current constraint
		addConstraint(m, lij[i] * nus[i] >= P[i]^2 + Q[i]^2);

	end

	# addConstraint(m, deltasSum == N-M);

	# add cuts
	for i = 1:N
		voltageEffectByLoadI = 2 * nEdgesOnPath[i] * (r * PCD[i] + x * QCD[i]);
		for j = 1:N
			if ancestors[i,j] == 1
				voltageEffectByLoadJ = 2 * nEdgesOnPath[j] * (r * PCD[j] + x * QCD[j]);
				if voltageEffectByLoadJ <= voltageEffectByLoadI
					addConstraint(m, gammas[i] <= gammas[j]);
				end
			end
		end
	end

	# define objective
	obj = 0;
	for i=1:N
		obj += Cll * (1-gammas[i]) * PCD[i] + r*lij[i]
	end
	obj += Clovr*t

	[m deltas gammas t obj nus P Q pc qc pg qg]
end


function countCommonEdges(par)
	paths = Array{Int64,2}[];
	for i = 1:N
		path = Int64[];
		node = i;
		while node != 0;
			push!(path,node);
			node = par[node];
		end
		push!(paths, path');
	end

	commonEdgesList = Array{Int64,2}[];
	for i = 1:N
		commonEdges = zeros(Int64, N);
		p1 = paths[i];
		

		for j = 1:N
			k = 0;	
			p2 = paths[j];
            l1 = length(p1);
			l2 = length(p2);
			while l1 > 0 && l2 > 0 && p1[l1] == p2[l2]
				k += 1;
				l1 -= 1;
				l2 -= 1;
			end
            commonEdges[j] = k;
		end
        push!(commonEdgesList, commonEdges');
	end

    commonEdgesList
end
commonEdgesList = countCommonEdges(par)

objValue = -100Clovr;
lovrValues = Float64[];
vollValues = Float64[];
targetNodesList = Array{Int64,1}[];
nnus = zeros(Float64, N);

for M = 3:3
	allNodes = [1:N];
	for comb in combinations(allNodes, M)

		# initialize ngammas
		targetNodes = zeros(Int64, M);
		ngammas = ones(Float64, N);
		currentBestObjValue = -100Clovr;
		finalObjValue = -100Clovr;
		ndeltas = ones(Int64,N);
		maxIterations = 3;
		lovrValue = 0;
		vollValue = 0;	
		
		################################################################
		# Slave problem 
		################################################################
		mm = getNonLinearModel();
		m = mm[1]; 
		deltas = mm[2];
		gammas = mm[3];
		t = mm[4];
		obj = mm[5];
		nus = mm[6];
		P = mm[7];
		Q = mm[8];
		pc = mm[9];
		qc = mm[10];
		pg = mm[11];
		qg = mm[12];
		
		# if slave problem, then maximize the rewards for attacker
		@setObjective(m, :Min, obj)		
				
		# pivot node voltage constraints
		addConstraint(m, t >= 0);
		for i = 1:N
			addConstraint(m, t >= nusL - nus[i]);
		end	
		
		for i in comb	
			addConstraint(m, deltas[i] == 0);
			ndeltas[i] = 0;
		end

		solve(m)

		ngammas = zeros(Float64, N);
		for i=1:N
		    ngammas[i] = getValue(gammas[i]);
		end

		nnus = zeros(Float64, N);
		for i=1:N
		    nnus[i] = getValue(nus[i]);
		end

		finalObjValue = getObjectiveValue(m);
		if finalObjValue >= currentBestObjValue
			# update values
			currentBestObjValue = finalObjValue;
			lovrValue = Clovr * getValue(t);
	        vollValue = finalObjValue - lovrValue;	        
			
			j = 1;
			for i=1:N
			    if ndeltas[i] == 0
			    	targetNodes[j] = i;
			    	j += 1;
			    end
			end
		end

	end
	push!(lovrValues, lovrValue);
    push!(vollValues, vollValue);
    push!(targetNodesList, targetNodes);

end
```
