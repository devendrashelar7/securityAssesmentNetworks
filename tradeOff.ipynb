```julia
using JuMP
using Gurobi

midNode = 2;
#number of nodes. substation node not included
N = 3 * midNode
# number of DGs that can be compromised
M = 2;

m = Model();

# Binary Variables
@defVar(m, deltas[1:N], Bin)

# continous variables
@defVar(m, gammas[1:N] >= 0)

# voltage and current magnitude squares
@defVar(m, nus[1:N] >= 0)


# real and reactive and complex loads and DGs
@defVar(m, pc[1:N] >= 0)
@defVar(m, qc[1:N] >= 0)
@defVar(m, pg[1:N] >= 0)
@defVar(m, qg[1:N])


# real and reactive power flows
@defVar(m, P[1:N] >=0)
@defVar(m, Q[1:N])

#auxiliary variable
@defVar(m, t >= 0)

# power parameters
PCD = 1/10N
QCD = PCD/3

# line impedance parameters
r = 0.01
x = 0.01

PGM = PCD
SGM = PGM  
PGSP = SGM*r/(r^2+x^2)^0.5
QGSP = SGM*x/(r^2+x^2)^0.5

nus_0 = 1

# cost of load shedding
C = 1000
W = 10000000

# define objective
obj = 0;
for i=1:N
    obj += C*(1-gammas[i])
end
@setObjective(m, :Min, obj)

obj += W*t

# Constraints
VL = 0.97^2;
VU = 1.03^2;

deltasSum = 0;
addConstraint(m, nus[1] == nus_0 - 2(r * P[1] + x * Q[1]));

for i=1:N

    # Model constraints

    # Voltage constraints
    # addConstraint(m, VL <= nus[i] );
    # addConstraint(m, nus[i] <= VU );
    
    addConstraint(m, t >= VL - nus[i]);
    
    addConstraint(m, PCD*gammas[i] == pc[i] );
    addConstraint(m, QCD*gammas[i] == qc[i] );
    addConstraint(m, gammas[i] <= 1 );

    addConstraint(m, pg[i] == PGSP * deltas[i] ) ;
    addConstraint(m, qg[i] == -SGM + (QGSP + SGM) * deltas[i] );
    

    if (i >= 2 && i != midNode ) 
        addConstraint(m, nus[i] == nus[i-1] - 2 * (r * P[i] + x * Q[i]) );
    end

    # sums
    deltasSum += deltas[i];
    PSum = pc[i] - pg[i];
    QSum = qc[i] - qg[i];
    
    if (i % midNode != 0)
        PSum += P[i+1];
        QSum += Q[i+1];
    end 


    if (i != midNode ) 
        addConstraint(m, P[i] == PSum);
        addConstraint(m, Q[i] == QSum); 
    end
 
end

#resource constraint of the attacker
addConstraint(m, deltasSum == N - M );

#special fork constraints
addConstraint(m, nus[midNode+1] == nus[midNode] - 2(r * P[midNode+1] + x * Q[midNode+1]));
addConstraint(m, nus[2midNode+1] == nus[midNode] - 2(r * P[2midNode+1] + x * Q[2midNode+1]));
addConstraint(m, P[midNode] == pc[midNode] - pg[midNode] + P[midNode+1] + P[2midNode+1]);
addConstraint(m, Q[midNode] == qc[midNode] - qg[midNode] + Q[midNode+1] + Q[2midNode+1]);


n1 = 1
n2 = 2
#n3 = 5
#n4 = 6

addConstraint(m, deltas[n1] == 0);
addConstraint(m, deltas[n2] == 0); 
#addConstraint(m, deltas[n3] == 0);
#addConstraint(m, deltas[n4] == 0);

m

status = solve(m)

```
